# Verification of density profile with MG profile

MG profile denotes the Miyamoto-Nagai (1975) profile. Ref: Publ. Astron. Soc. Japan 27, 533-543 (1975). The potential is of the form:
\begin{equation}
\Phi(R, z) = \frac{GM}{\sqrt{R^2 + \left[a + \sqrt{z^2 + b^2} \right]^2}},
\end{equation}
where $M$ denotes the total mass of the disk, $a$ is the radial scale and $b$ is the vertical scale.
The mass density is given by:
\begin{equation}
            \rho(R, z) = \frac{b^2 M}{4 \pi} \left[\frac{aR^2 + \left(a + 3\sqrt{z^2 + b^2}\right)\left(a + \sqrt{z^2 + b^2}\right)^2}{\left[R^2 + \left(a + \sqrt{z^2 + b^2} \right)^2\right]^\frac{5}{2} (z^2 + b^2)^\frac{3}{2}}\right]
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy import units as u
from astropy import constants as const
from modules import TimingTools as TT
```

## Reading isolated sim data

```python
phase = Table.read('./data/LMC_iso_rotated_phase_space.fits')
```

## Reading scale lengths obtained from density profile fitting

```python
logrho0, e_logrho0, H, e_H, z0, e_z0 = np.load('./data/density_profile_params.npy')

H = H*u.kpc
z0 = z0*u.kpc
```

## Verification

```python
#Total mass of the disk
m = 2500*u.Msun #mass of a star particle
N = len(phase) #total number of star particles
M = m*N

r = np.linspace(0, 10, 100)*u.kpc
z = np.linspace(0, 1.5, 100)*u.kpc
```

### Radial density profile

```python
logrhor_actual = TT.log_rad_density(r, logrho0, H)
logrhor_MG = np.log10(TT.MG_rho(r, 0*u.kpc, H, z0, M).value)
```

```python
plt.figure(figsize = (10,6))

plt.plot(r, logrhor_actual, label = 'Actual density profile')
plt.plot(r, logrhor_MG, label = 'MG disk')
plt.xlabel('r (kpc)')
plt.ylabel(r'$log(\rho(r))$ ($M_\odot$ kpc$^{-3}$)')
plt.legend()
plt.show()
```

### Vertical density profile

```python
logrhoz_actual = TT.log_vert_density(z.value, logrho0, z0.value)
logrhoz_MG = np.log10(TT.MG_rho(0*u.kpc, z, H, z0, M).value)
```

```python
plt.figure(figsize = (10,6))

plt.plot(z, logrhoz_actual, label = 'Actual density profile')
plt.plot(z, logrhoz_MG, label = 'MG disk')
plt.xlabel('z (kpc)')
plt.ylabel(r'$log(\rho(z))$ ($M_\odot$ kpc$^{-3}$)')
plt.legend()
plt.show()
```
