# Comparison of MN and Exponential profile parameters

```python
import numpy as np
import getdist
from getdist import MCSamples, plots
from getdist.chains import Chains
import matplotlib as mpl
import matplotlib.pyplot as plt 
from modules import potential_tools as pt
import copy
import matplotlib as mpl
```

```python
mpl.rcParams['axes.labelsize'] = 'xx-large'
mpl.rcParams['xtick.labelsize'] = 'xx-large'
mpl.rcParams['ytick.labelsize'] = 'xx-large'
```

## Chains

```python
samples_exp = np.loadtxt('./chains/mcmc_exp_profile.txt')
samples_mn = np.loadtxt('./chains/mcmc_MN_profile.txt')
```

## Contours

```python
getdist_samples = []
getdist_samples.append(MCSamples(samples = samples_exp, names = ['a', 'b', 'logM'], label = 'Exponential fit'))
getdist_samples.append(MCSamples(samples = samples_mn, names = ['a', 'b', 'logM'], label = 'Miyamoto - Nagai fit'))
```

```python
g = plots.get_subplot_plotter()
g.triangle_plot(getdist_samples, filled = True)

ax_a_b = g.get_axes_for_params('a', 'b')
ax_a_b.set_ylabel('b (kpc)')

ax_a_logM = g.get_axes_for_params('a', 'logM')
ax_a_logM.set_xlabel('a (kpc)')
ax_a_logM.set_ylabel(r'$log(M/M_\odot)$')

ax_b_logM = g.get_axes_for_params('b', 'logM')
ax_b_logM.set_xlabel('b (kpc)')

ax_logM_logM = g.get_axes_for_params('logM')
ax_logM_logM.set_xlabel(r'$log(M/M_\odot)$')

plt.savefig('./figures/mn_exp_contours.pdf', bbox_inches = 'tight')

plt.show()
```
